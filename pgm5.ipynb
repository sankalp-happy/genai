```python
import gensim.downloader as api
w2v=api.load("word2vec-google-news-300")
print('word2vec loaded')
```

```python
from transformers import pipeline
generator=pipeline('text-generation', model='gpt2')
print('gpt2 loaded')
```

```python
seed_words=input('enter the seed words: ')
print('your seed words are: ', seed_words)
```

```python
similar=w2v.most_similar(seed_words, topn=5)
similar_words=[word for word, _ in similar]
print('similar words to seed words: ', similar_words)
if similar_words:
    prompt_fot_gpt2=f'write a short creative pra about {seed_words} using the following similar words: {",".join(similar_words)}. enriched prompt: '
    result=generator(prompt_fot_gpt2, max_length=100, num_return_sequences=1,do_sample=True, temperature=0.9)
    generated_text=result[0]['generated_text']
    print('='*60)
    print('seed_words',seed_words)
    print('similar_words',similar_words)
    print('generated_text',generated_text)
    print('='*60)
else:
    print('no similar words found for the seed words')
```
