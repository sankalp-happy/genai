```python
import nltk
import pandas as pd
from gensim.models import Word2Vec
nltk.download("stopwords")
```

```python
sentences , domains = [], []
for line in open("corpus.txt", "r"):
    line = line.strip()
    if line in ["MEDICAL", "LEGAL"]:
        domain = line.lower()
    else:
        sentences.append(line.split())
        domains.append(domain if 'domain' in locals() else None)
df = pd.DataFrame({"sentence": sentences, "domain": domains})

print(f"Total sentences: {len(df)}")
print(f"Medical sentences: {len(df[df['domain'] == 'medical'])}")
print(f"Legal sentences: {len(df[df['domain'] == 'legal'])}")
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))

corpus = [
    [w for w in sentence.lower().split() if w not in stop_words and len(w) > 2]
    for sentence in df['sentence']
]

corpus = [tokens for tokens in corpus if tokens]
print(f"Total sentences after preprocessing: {len(corpus)}")
print(f"Before: {df['sentence'][0]}")
print(f"After: {corpus[0]}")
```
