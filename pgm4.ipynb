```python
import gensim.downloader as api
w2v=api.load("word2vec-google-news-300")
print('word2vec loaded')
```

```python
from transformers import pipeline
generator=pipeline('text-generation', model='gpt2')
print('gpt2 loaded')
```

```python
original_prompt=input('Enter a prompt: ')
key_words=input('enter the key words: ')
print('your prompt is: ', original_prompt)
print('your key words are: ', key_words)
```

```python
similar=w2v.most_similar(key_words, topn=5)
similar_words=[word for word, _ in similar]
print('similar words to key words: ', similar_words)
```

```python
if similar_words:
    prompt_for_gpt2 = (
        f"rewrite this prompt in a more descriptive and enriched way "
        f"original prompt: {original_prompt} "
        f"{', '.join(similar_words)} "
        f"enriched prompt: "
    )
    result = generator(
        prompt_for_gpt2,
        max_length=100,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.9,
    )
    generated_text = result[0]["generated_text"]
    print("=" * 60)
    print("original_prompt", original_prompt)
    print("enriched_prompt", similar_words)
    print("generated_text", generated_text)
    print("=" * 60)
else:
    print("could not enrich.please try with different key words")
```
